```python
import numpy as np
import matplotlib.pyplot as plt
import sklearn.metrics as sk_metrics

fpr_proposed = np.loadtxt("fprs_proposed.txt")
tpr_proposed = np.loadtxt("tprs_proposed.txt")
auc_proposed = sk_metrics.auc(fpr_proposed, tpr_proposed)

fpr_odin = np.loadtxt("fprs_odin.txt")
tpr_odin = np.loadtxt("tprs_odin.txt")
auc_odin = sk_metrics.auc(fpr_odin, tpr_odin)

fpr_base = np.loadtxt("fprs_baseline.txt")
tpr_base = np.loadtxt("tprs_baseline.txt")
auc_base = sk_metrics.auc(fpr_base, tpr_base)
```

```python
idx_tpr_95_proposed = np.where(tpr_proposed >= 0.95)[0][0]
idx_tpr_95_odin = np.where(tpr_odin >= 0.95)[0][0]
idx_tpr_95_base = np.where(tpr_base >= 0.95)[0][0]
```

```python
fig, ax = plt.subplots()
lw = 2
plt.plot(fpr_proposed, tpr_proposed, color='green',
         lw=lw, label='Proposed (area = %0.2f)' % auc_proposed)


plt.plot([0, 1], [0, 1], color='gray', lw=1, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])


plt.plot(fpr_odin, tpr_odin, color='blue',
         lw=lw, label='ODIN (area = %0.2f)' % auc_odin)


plt.plot(fpr_base, tpr_base, color='red',
         lw=lw, label='Baseline (area = %0.2f)' % auc_base)

plt.grid(True)



# FPR at 95 TPR
ax.scatter(fpr_proposed[idx_tpr_95_proposed], tpr_proposed[idx_tpr_95_proposed], s=100, c="green")
ax.scatter(fpr_odin[idx_tpr_95_odin], tpr_odin[idx_tpr_95_odin], s=100, c="blue")
ax.scatter(fpr_base[idx_tpr_95_base], tpr_base[idx_tpr_95_base], s=100, c="red")

ax.annotate("TPR=95%", (fpr_proposed[idx_tpr_95_proposed]-0.07, tpr_proposed[idx_tpr_95_proposed]+0.05))
ax.annotate("TPR=95%", (fpr_odin[idx_tpr_95_odin]-0.07, tpr_odin[idx_tpr_95_odin]-0.08))
ax.annotate("TPR=95%", (fpr_base[idx_tpr_95_odin]-0.07, tpr_odin[idx_tpr_95_odin]-0.08))


plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc="lower right")
plt.show()
```
